```python
lqm = biolqm.load("http://ginsim.org/sites/default/files/SuppMat_Model_Master_Model.zginml")
```

```python
bn = biolqm.to_minibn(lqm)
#bn
```

```python
attractors = cabean.attractors(bn)
tabulate(attractors)
```

HS = attractor 3  
EMT = attractors 1 & 4  
DNAdamage is on in 1, and off in 3 and 4, thus we are interested in 3 to 4

# One-step strategies

```python
onestep = cabean.OneStepReprogramming(bn)
```

# Attractor based strategies

```python
attr= cabean.AttractorSequentialReprogramming(bn)
```

```python
attr.attractor_to_attractor(attractors[3],attractors[4])
```
