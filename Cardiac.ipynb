Tests on the cardiac network, from cellcollective to analysis with cabean

```python
import cellcollective
import biolqm
import cabean
from colomoto_jupyter import tabulate
import pandas as pd
from IPython.display import *
```

```python
cabean.enable_debug()
```

```python
model = cellcollective.load("https://cellcollective.org/#2136:1/cardiac-development/1")
```

```python
lqm = cellcollective.to_biolqm(model)
```

```python
bn = biolqm.to_minibn(lqm)
```

```python
attractors = cabean.attractors(bn)
tabulate(attractors)
```

Attractors :
<br>0 : Inactive
<br>1 : FHF muted
<br>2 : Active BMP
<br>3 : FHF
<br>4 : SHF muted
<br>5 : SHF
<br>We are looking for the reprogramming from SHF (=5) to FHF (=3)

# One-step reprogramming strategies

```python
onestep = cabean.OneStepReprogramming(bn)
```

```python
onestep.attractor_to_attractor(attractors[5],attractors[3])
```

# Attractor-based strategies

```python
attrbased= cabean.AttractorSequentialReprogramming(bn)
```

# Completely sequential strategies

```python
seq = cabean.SequentialReprogramming(bn)
```

```python
res = seq.attractor_to_attractor(attractors[5],attractors[3],maxsteps=2,limit=10000)
```
